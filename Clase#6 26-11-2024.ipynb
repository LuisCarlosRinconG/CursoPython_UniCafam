```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
comparacion=pd.read_csv('/content/sample_data/comparacion_ventas_exp.csv')
composicion=pd.read_csv('/content/sample_data/composicion_mercado_exp.csv')
correlacion=pd.read_csv('/content/sample_data/correlacion_marketing_exp.csv')
distribucion=pd.read_csv('/content/sample_data/distribucion_tiempos_exp.csv')
```

***Grafica de correlación***

```python
correlacion.head(5)
```

```python
# Grafica de correlación de la inversión vs las ventas generadas

# df=sns.load_dataset(correlacion)

sns.scatterplot(x=correlacion["Inversión en marketing (USD)"],y=correlacion['Ventas generadas (USD)'])
```

```python
sns.scatterplot(x=correlacion["Inversión en marketing (USD)"],y=correlacion['Ventas generadas (USD)'],color="red")
```

```python
sns.scatterplot(x=correlacion["Inversión en marketing (USD)"],
                y=correlacion['Ventas generadas (USD)'],
                hue=correlacion["Campaña ID"],legend=True)
```

```python
# Grafica heatmap

sns.heatmap(correlacion[["Inversión en marketing (USD)","Ventas generadas (USD)"]])
```
