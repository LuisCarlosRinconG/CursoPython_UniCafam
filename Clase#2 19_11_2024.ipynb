**OPERADORES MATEMATICOS **

```python
a = 2
b = 4

#suma
print(a+b)#6

#resta
print(b-a)#2

# División
c=b/a #2
print(c)

#División entera
d=3
print(d//a)#2

#Exponenciación
print(a**5)

#Modulo

print(b%a)

#Asignación compuesta

a+=b
print(a)
a=2

#Redondeo
e=3.14159
print(round(e,2))

#Valor absoluto se utiliza abs

f=a-b
print(abs(f))
```

**Textos (Valores string)**

```python
# Concatenación de caracteres

saludo='Hola'

nombre='mundo'

mensaje=saludo+' , '+nombre

print(mensaje)
```

```python
#Repetición de caracteres

repetido=saludo*3
print(repetido)
```

```python
#INDEXACIÓN

primer_caracter=saludo[0]
print(primer_caracter)

#extraer hola
subcadena=mensaje[0:4]
print(subcadena)
```

```python
#metodos de cadenas

#Mayusculas

cadena='Python es genial'

print(cadena.upper())

#minuscaulas

print(cadena.lower())

#remplazo

print(cadena.replace('genial','increible'))

#separar
print(cadena.split(" "))

print(cadena[0])

cadena_separado=cadena.split(" ")
print(cadena_separado[0])
```
